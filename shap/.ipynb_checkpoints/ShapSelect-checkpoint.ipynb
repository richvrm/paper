```python
import xgboost
import shap
import pandas as pd
import matplotlib.pyplot as plt
```

```python
shap.initjs()
```

```python
#importar dataset
dataset = pd.read_csv('../datasets/selectFeatures.csv',index_col=0)
```

```python
X = dataset
y = dataset.iloc[:,21].values #o que deve ser predito
```

```python
X = X.drop(columns=['results'])
```

```python
# train XGBoost model
#X,y = shap.datasets.boston()
model = xgboost.train({"learning_rate": 0.01}, xgboost.DMatrix(X, label=y), 100)
```

```python
y
```

```python
# explain the model's predictions using SHAP
# (same syntax works for LightGBM, CatBoost, scikit-learn and spark models)
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)
```

```python
# 0  adaptive
#1 corrective
#2 perfective
    # visualize the first prediction's explanation (use matplotlib=True to avoid Javascript)
shap.force_plot(explainer.expected_value, shap_values[0,:], X.iloc[0,:],  matplotlib=False)
```

```python
# visualize the training set predictions
shap.force_plot(explainer.expected_value, shap_values, X)
```

```python
# create a dependence plot to show the effect of a single feature across the whole dataset
shap.dependence_plot("remov", shap_values, X)
```

```python
# summarize the effects of all the features
shap.summary_plot(shap_values, X)
```

```python
shap.summary_plot(shap_values, X, plot_type="bar",color="red")
```
